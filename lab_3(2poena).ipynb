```python
# Отстрани редици кои имаат '-' во која било од колоните
X_train = X_train[X_train != '-'].dropna()
Y_train = Y_train[Y_train != '-'].dropna()

# Ако се уште има податоци, можеш да примениш StandardScaler
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)

# Применете го истиот процес на X_test ако е потребно
X_test = X_test[X_test != '-'].dropna()
X_test = scaler.transform(X_test)
```

```python
X=data.drop('PriceEuro',axis=1)
Y=data['PriceEuro']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor

# Инициализирање на моделот DecisionTreeRegressor со зададени параметри
model = DecisionTreeRegressor(max_depth=5, min_samples_split=5, max_features=0.5)

# Обука на моделот со функцијата fit
model.fit(X_train, Y_train)

# Провери го моделот (можеш да го користиш за предвидување или оценување)
y_pred = model.predict(X_test)

# Ако сакаше да ја процениш точноста, можеш да го направиш тоа:
from sklearn.metrics import mean_squared_error,  r2_score, mean_absolute_error
mse = mean_squared_error(Y_test, y_pred)
print(f"Mean Squared Error: {mse}")
print(mean_absolute_error(Y_test, y_pred), mean_squared_error(Y_test,y_pred), r2_score(Y_test,y_pred))
```

```python
from xgboost import XGBRegressor

# Инициализација на моделот со зададени параметри
model = XGBRegressor(
    n_estimators=100,         # Број на дрвја во ансамблот
    max_depth=5,              # Максимална длабочина на дрвјата
    learning_rate=0.1,        # Стапка на учење (learning rate)
    objective="reg:squarederror"  # Функција на целта за регресија
)

# Обука на моделот со функцијата fit
model.fit(X_train, Y_train)

# Пример за предвидување
y_pred = model.predict(X_test)
print(mean_absolute_error(Y_test, y_pred), mean_squared_error(Y_test,y_pred), r2_score(Y_test,y_pred))

```
