```python

import pandas as pd
```

```python
data = pd.read_csv('/Users/Korisnik55/ElectricCarData.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoded_brand = encoder.fit_transform(data['Brand'])
data['Brand'] = encoded_brand
encoded_model = encoder.fit_transform(data['Model'])
data['Model'] = encoded_model
encoded_charge = encoder.fit_transform(data['RapidCharge'])
data['RapidCharge'] = encoded_charge
encoded_train = encoder.fit_transform(data['PowerTrain'])
data['PowerTrain'] = encoded_train
encoded_type = encoder.fit_transform(data['PlugType'])
data['PlugType'] = encoded_type
encoded_style = encoder.fit_transform(data['BodyStyle'])
data['BodyStyle'] = encoded_style
encoded_segment = encoder.fit_transform(data['Segment'])
data['Segment'] = encoded_segment
```

```python
data
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(max_depth=3,min_samples_leaf=5,max_leaf_nodes=10,min_samples_split=5)
```

```python
data
```

```python
import numpy as np
data = data.replace('-', np.nan)
```

```python
data
```

```python
count_dash = data.apply(lambda col: col.isin(['-']).sum())
print(count_dash)
```

```python
X = data.drop(['PriceEuro'],axis=1)
Y = data['PriceEuro']
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
from sklearn.metrics import r2_score
r2_score(Y_test,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(Y_test, y_pred)
```

```python
mae
```

```python
X_train['FastCharge_KmH'] = pd.to_numeric(X_train['FastCharge_KmH'], errors='coerce')
X_test['FastCharge_KmH'] = pd.to_numeric(X_test['FastCharge_KmH'], errors='coerce')
```

```python
from xgboost import XGBRegressor

# Initialize the XGBRegressor
xgb_model = XGBRegressor(max_depth=50, min_child_weight=1, n_estimators=200, 
                         n_jobs=-1, verbose=1, learning_rate=0.18)

# Fit the model
xgb_model.fit(X_train, Y_train)
```

```python
y_pred = xgb_model.predict(X_test)
```

```python
r2_score(Y_test,y_pred)
```

```python
mae = mean_absolute_error(Y_test, y_pred)
```
